```python
# para fazer leitura de uma página da web, o primeiro passo é importar a blibioteca request
import requests
url = "https://www.worldcoinindex.com"
# Usando a função get e passando o site adquirimos uma resposta, que contém HTML do site
resposta = requests.get(url)
resposta
```

```python
# podemos usar a função do pandas rad para ler o conteudo HTML de dentro da respostas
leitura_html = pd.read_html(resposta.text)
leitura_html
```

```python
print("Tipo = ", type(leitura_html))
# neste caso, é uma lista
```

```python
# neste caso, é uma lista com um elemento só
print("tamanho =", len(leitura_html))
```

```python
# Então vamos verificar o tipo desse elemento
print("TipoDoElemento =", type(leitura_html[0]))
# é um dataframe
```

```python
tabela = leitura_html[0]
tabela
```

```python
tabela.head()
# Mostra as 5 primeiras loinhas da tabela
```

```python
# para remover essas colunas usamos o loc
# No primeiro atributo dezemos que linhas queremos localizar, usamis : para todas
# No segundo dizemos que colunas. Nesse caso selecionamos da coluna Nome para frente
# colocariamos o : no final, caso desejassemos selecionar as primeiras colunas até nome
tabela = tabela.loc[:,"Name":]
tabela
```

```python
# podemosusar a função dropna para remover dados faltantes
# A função recebe um parametro chamado axis.
# Quando esse parametro é igual a 0, significa que estamos buscando linhas com elementos vazios
# Quanto esse paramentro é igual 1, significa que estamos buscando colunas com elementos vazios
tabela.dropna(axis=1)
```

Salvar arquivo como excel

```python
# primeiro voce cria o writer, aquele que escreve o arquivo. Passando como parametro o nome do arquivo
# e qual programa voce vai usar para gerar o arquivo.
# Para excel a engine é xlsxwriter
writer = pd.ExcelWriter("imoveis-alterado.xlsx", engine="xlsxwriter")
# depois disso, use o writer para converter as tabelas nesse arquivo
imoveis.to_excel(writer, sheet_name="imoveis")
# Por fim, chame a função save, para salvar o arquivo
writer.save()
```
