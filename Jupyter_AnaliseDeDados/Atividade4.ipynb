```python
import pandas as pd
```

```python
dados = pd.read_csv("resultadosfinais2017.csv", sep=";")
dados
```

```python
# Verifique quais os bairros possuem escolas no conjunto de dados.
dados["ebairrnome"].unique()
```

```python
# Descubra o número de alunos por bairro
dados["ebairrnome"].value_counts()
```

```python
dados["esitaldesc"].value_counts()
```

```python
# Descubra o número de alunos reprovados por escola.
reprovados = dados[dados["esitaldesc"] != "APROVADO"]
reprovados["nescolnome"].value_counts()
```

```python
# Descubra o número de alunos aprovados por bairro. 
aprovados_bairros = dados[dados["esitaldesc"] == "APROVADO"]
aprovados_bairros["ebairrnome"].value_counts()
```

```python
# Descubra o número de alunos reprovados que estudam pela MANHÃ em cada escola.
reprovados_manha = reprovados[reprovados["nturesnome"] == "MANHÃ"]
reprovados_manha["nescolnome"].value_counts()
```

```python
# Descubra o número de alunos reprovados que estudam pela NOITE em cada Bairro
reprovados_noite = reprovados[reprovados["nturesnome"] == "NOITE"]
reprovados_noite["ebairrnome"].value_counts()
```
